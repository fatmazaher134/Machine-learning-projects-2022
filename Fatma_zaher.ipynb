```python
# reading the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('C://Users//Nmber One//Documents//4_5901974192206645743.csv')
data.head()
```

```python
data.replace('',np.nan)
```

```python
# convert string to a number
from sklearn import preprocessing
obj=data.select_dtypes(include=['object'])
le=preprocessing.LabelEncoder()
for i in range(obj.shape[1]):
    obj.iloc[:,i]=le.fit_transform(obj.iloc[:,i])
nun_obj=data.select_dtypes(exclude=['object'])
data=pd.concat([obj,nun_obj],axis=1)
```

```python
# handel null for the coulmn rbc
for i  in range (data['rbc'].count()):
    if data['rbc'][i]=='normal':
        data['rbc'][i]='normal'
    elif data['rbc'][i]=='abnormal':
        data['rbc'][i]='abnormal'
    else:
        data['rbc'][i]=data['rbc'].mode()[0]
```

```python
# handle null for the coulmns
for i  in  (data.columns):
    if data[i].isnull:
        data[i]=data[i].fillna(data[i].mode()[0])
```

```python
# split the data to x , y and train , test
x=data.drop(['id','classification'],axis=1)
y=data['classification']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x
```

```python
# create the model
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators=100,max_depth=10)
model.fit(x_train,y_train)

pred=model.predict(x_test)
print(model.score(x_test,y_test))
```
